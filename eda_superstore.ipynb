## Data Transformation

```python
# Ekstracting atributes for order date - Year and Month
df_orders['order_date_year'] = df_orders['Order Date'].apply(lambda x: x.year)
df_orders['order_date_month'] = df_orders['Order Date'].apply(lambda x: x.month)
df_orders['order_date_month_name'] = df_orders['Order Date'].apply(lambda x: x.strftime('%b'))
df_orders['order_date_year_month'] = df_orders['Order Date'].apply(lambda x: x.strftime('%Y%m'))
df_orders['order_date'] = df_orders['Order Date'].apply(lambda x: x.strftime('%Y%m%d'))

# Ekstracting attributes for order date - Day and Day of week
df_orders['order_date_day'] = df_orders['Order Date'].apply(lambda x: x.day)
df_orders['order_date_dayofweek'] = df_orders['Order Date'].apply(lambda x: x.dayofweek)
df_orders['order_date_dayofweek_name'] = df_orders['Order Date'].apply(lambda x: x.strftime('%a'))

# Create Quarter column
def quarter(x):
  if x <= 3:
    return 'Q1'
  elif x > 3 and x <= 6:
    return 'Q2'
  elif x > 6 and x <= 9:
    return 'Q3'
  else:
    return 'Q4'

df_orders['Quarter'] = df_orders['order_date_month'].apply(quarter)

# New Orders Dataframe after transformations
df_orders.head()
```

## Exploratory Data Analysis (EDA)
###

```python
# Create new dataframe Actual vs Target
df_orders_target = df_orders.groupby(by='order_date_year_month', as_index=False).sum().loc[:,['order_date_year_month','Sales']]
df_orders_target.rename(columns={'Sales': 'Actual'}, inplace=True)

df_orders_target['Target'] = df_orders_target['Actual'].apply(lambda x: x + (x*0.1))

df_orders_target
```

```python
def segment(x):
  if x <= 200:
    return 'Bronze'
  elif x > 200 and x < 500:
    return 'Silver'
  else:
    return 'Gold'

# segmentasi Customer by Sales per Quarter
df_orders_customer_segments = df_orders.groupby(by=['order_date_year','Quarter','Customer ID'], as_index=False).sum().loc[:,['Quarter','order_date_year','Customer ID','Sales']]
df_orders_customer_segments['Segment'] = df_orders_customer_segments['Sales'].apply(segment)

df_orders_customer_segments
```

```python
def state_segment(x):
  if x <= 1000:
    return 'Kategori I'
  elif x > 1000 and x < 2000:
    return 'Kategori II'
  else:
    return 'Kategori III'

# segmentasi toko cabang
df_orders_state = df_orders.groupby(by=['State','order_date_year_month'], as_index=False).sum().loc[:,['State','order_date_year_month','Sales']]
df_orders_state['Segment'] = df_orders_state['Sales'].apply(state_segment)

df_orders_state
```

```python
def product_segment(x):
  if x <= 5:
    return '3rd Product'
  elif x > 5 and x < 10:
    return '2nd Product'
  else:
    return '1st Product'

# segmentasi produk dengan penjualan
df_orders_product = df_orders.groupby(by=['Region', 'State', 'order_date_year_month', 'Product ID'], as_index=False).count().loc[:,['Region','State','order_date_year_month','Product ID','Order ID']]
df_orders_product['Segment'] = df_orders_product['Order ID'].apply(product_segment)

df_orders_product
```

```python
df_product = df_orders_product.groupby(by=['Product ID'], as_index=False).count().loc[:,['Product ID','Order ID']]
df_product
```
